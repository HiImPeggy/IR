```python
import json

# 读取.json文件
with open('/content/drive/MyDrive/Colab Notebooks/wiki_article_list_2023_tra.json', 'r', encoding='utf-8') as f:
    data = json.load(f)


# 构建输出模式并写入新的.json文件
output_data = []
for i, item in enumerate(data):
    output_data.append({"id": f"doc{i+1}", "contents": item})

# 输出为新的.json文件
with open('/content/IR/corups/retriver.json', 'w', encoding='utf-8') as f:
    json.dump(output_data, f, ensure_ascii=False, indent=2)

```

```python
from pyserini.index import IndexReader
from math import log
from pyserini.search.lucene import LuceneSearcher
```

```python
index_reader = IndexReader('indexes/sample_collection_jsonl_zh')
total_docs = index_reader.stats().get('documents')
print(total_docs)
```

```python
searcher = LuceneSearcher('indexes/sample_collection_jsonl_zh')
searcher.set_language('zh')
hits = searcher.search('蝙蝠俠',k=total_docs)
print(len(hits))
```

```python
idf = log(total_docs / (len(hits)))
print(f'IDF 值為：{idf}')
```
